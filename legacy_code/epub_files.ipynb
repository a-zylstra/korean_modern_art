```python
import re
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup as bs
import pandas as pd
from pathlib import Path
```

# Converting Epub into Txt

```python

for path in Path("unformatted_texts/").glob("*.epub"):
    book = epub.read_epub(path)
    doc_content = book.get_items_of_type(ebooklib.ITEM_DOCUMENT)
    full_text = ""

    for content in doc_content:
        soup = bs(content.get_content(), "xml")
        text = soup.get_text()
        full_text += text
    
    title = book.get_metadata("DC", "title")[0][0]
    year_replace = title.replace('â€“',' to ')
    cleaned = re.sub(r"[^\w\s]", "", year_replace)
    filename = "plain_text/" + re.sub(r"\s+", "_", cleaned).strip().lower() + '.txt'
    with open(filename, "w", encoding='utf-8') as f:
        f.write(full_text)
    print(f'original file name: {path}')
    print(f'new file name: {filename}')
```
