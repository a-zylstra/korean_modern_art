```python
import re
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup as bs
import pandas as pd
from pathlib import Path
```

```python
title_pattern = re.compile(r"^Title:\s*(.*?)\s*Author:", re.IGNORECASE | re.DOTALL)
# summation_pattern = re.compile(r'SUMMATION.*?\d{4}', re.IGNORECASE | re.DOTALL)
summation_pattern = re.compile(r'SUMMATION.{0,70}?\s+\d{1,2}\s+\w+\s+\d{4}', re.IGNORECASE | re.DOTALL)


for path in Path("unformatted_texts/").glob("*.txt"):
    text = path.read_text(encoding="utf-8")
    # getting titles for file names 
    titles = title_pattern.findall(text)
    if titles:
        for x in titles:
            year_replace = x.replace('-',' to ')
            cleaned = re.sub(r"[^\w\s]", "", year_replace)
            filename = "plain_text/" + re.sub(r"\s+", "_", cleaned).strip().lower() + '.txt'
    else:
        match = summation_pattern.search(text)
        if match:
            x = match.group(0)
            cleaned = re.sub(r"[^\w\s]", "", x)
            filename = "plain_text/" + re.sub(r"\s+", "_", cleaned).strip().lower() + '.txt'
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(text)
    print(f'New filename: {filename}')
    print(f'Original file name: {path}')

```
