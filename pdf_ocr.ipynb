```python
import os
from pdf2image import convert_from_path
```

```python
file_name = "art_of_dansaekhwa_press_release"

images = convert_from_path("raw_files/art_of_dansaekhwa_press_release.pdf"
                           , fmt="tiff"
                           , dpi=400
                           , poppler_path = r"C:\Program Files\poppler-25.12.0\Library\bin"
                           )

os.mkdir("images")
page_count = 1
for image in images:
    page_image_name = f"images/{file_name}-page-{page_count}.tiff"
    image.save(page_image_name, format="tiff")
    page_count += 1
```
