```python
import os
from pdf2image import convert_from_path
from pathlib import Path
import pytesseract
from bs4 import BeautifulSoup as bs

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract"
```

```python
folder_name = "raw_files"
for path in Path(f"{folder_name}").glob("*.pdf"):
    images = convert_from_path(f"{path}"
                            , fmt="tiff"
                            , dpi=400
                            , poppler_path = r"C:\Program Files\poppler-25.12.0\Library\bin"
                            )
    
    file_name = f"{str(path).replace(f'{folder_name}','').replace(".pdf","")}"
    page_count = 1
    for image in images:
        page_image_name = f"images/{file_name}-page-{page_count}.tiff"
        image.save(page_image_name, format="tiff")
        print(f"{page_image_name}")
        page_count += 1
        
    page_images = range(1, page_count)

    with open(f"plain_text/{file_name}.txt", "w", encoding="utf8") as output:
        for page in page_images:
            page_path = f"images/{file_name}-page-{page}.tiff"
            print(f"Processing {page_path}")
            xml = pytesseract.image_to_alto_xml(page_path)
            soup = bs(xml, "xml")
            text_blocks = soup.find_all("TextBlock")
            for text_block in text_blocks:
                text_lines = text_block.find_all("TextLine")
                paragraph = []
                for text_line in text_lines:
                    string_tags = text_line.find_all("String")
                    line = []
                    for string_tag in string_tags:
                        line.append(string_tag["CONTENT"])
                    line_of_text = " ".join(line)
                    paragraph.append(line_of_text)
                paragraph_text = "\n".join(paragraph)
                output.write(f"{paragraph_text}\n\n")
            output.write("\n\n")        
```
